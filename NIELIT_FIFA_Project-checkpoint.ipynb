## Data Cleaning

```python
df.drop(['Unnamed: 0'], axis =1,inplace = True)
#df.head()
```

```python
df.drop(['Photo','Flag'],axis =1,inplace = True)
#df.head()
```

```python
df.drop(['Club Logo'],axis = 1,inplace = True)
```

```python
df.head(10).style.background_gradient('PuBu')
```

## Data Analysis

### Top 10 Costliest Players

```python
df['Wage'] = df['Wage'].replace({'€':'','K': '*1e3', 'M': '*1e6'}, regex=True).fillna(1).map(pd.eval).astype(int)
```

```python
sorted_top10 = df.sort_values('Wage',ascending = False).head(10)
sorted_top10[['Name','Age','Nationality','Club','Wage']].style.background_gradient('copper')
```

### Indian Players 

```python
df[df['Nationality']=='India'][['Name','Age','Position','Overall']].style.background_gradient('magma')
```

### Best Players in Overall Rating

```python
df.sort_values(['Overall','Potential'],ascending = False)[['Name','Age','Nationality','Club','Overall','Potential']].head(10).style.background_gradient('viridis')
```

### Count of number of players according to Positions  

```python
value_counts_Position = df['Position'].value_counts()
#value_counts_Position
```

```python
value_counts_Position.plot(kind = 'bar',width = .5,color = ['red','blue','green','orange'])
plt.suptitle('Players count according to Positions', color='w')
plt.show()
```

### Number of Clubs 

```python
df['Club'].value_counts().count()
```

### Number of players of Each country

```python
player_count = df['Nationality'].value_counts().head(10)
#player_count
```

```python
#player_count.plot(kind = 'pie')
#plt.show()
player_count.plot.pie(autopct="%.1f%%");
plt.suptitle('', color='w')
plt.show()
```

### Coutries giving MAX Total Wage 

```python
wage_nationality = df.groupby('Nationality')['Wage'].sum()
wage_nationality_sort = wage_nationality.sort_values(ascending = False).head(10)
wage_nationality_sort = pd.DataFrame(wage_nationality_sort)
wage_nationality_sort.style.background_gradient('PuBu')
```

### Max and Min Number of Players in a Club

```python
ma = max(df['Club'].value_counts())
#ma
```

```python
mi = min(df['Club'].value_counts())
#mi
```

```python
plt.bar(x = ['Min','Max'], height = [mi,ma], width =.3, color =['grey','brown'])
plt.title('Max and Min Varience')
plt.suptitle('', color='w')
plt.show()
```

### Players with different Foot Preference

```python
left = df.loc[df['Preferred Foot']=='Left'].count()['ID']
#left
```

```python
right = df.loc[df['Preferred Foot']=='Right'].count()['ID']
#right
```

```python
plt.bar(x = ['Left','Right'], height = [left,right], width =.3, color =['cyan','pink'])
plt.title('Players with different Foot Preference')
plt.suptitle('', color='w')
plt.show()
```
